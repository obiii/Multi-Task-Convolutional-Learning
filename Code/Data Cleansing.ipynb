```python
import os
import pandas as pd
import numpy as np
data = pd.read_csv('../CSV/flameData.csv')
data = data.iloc[:,2:].copy()
print('Data Shape: ',data.shape)
print(data.head())
print (data.dtypes)
```

If etylene flow is greater than 0.1, the fuel is ethane. (Proved: I have check Nan count)

```python
cols = ['PFR','H2Mol','N2Mol','NGMol','MBP10_CF002','MBP10_CF003','MBP10_CF004','EtylenFlow','H2Weight','NGWeight','N2Weight']
data[cols] = data[cols].clip(lower = 0)
data[cols] = np.round(data[cols])
data[cols]
```

## PFR distribution

```python
d = pd.DataFrame(data['PFR'].value_counts())
d['PFR_Val'] = d.index
d.sort_values(by=['PFR_Val'])
```

```python
def labelObs(row):
    if row['PFR'] == 0:
        return '0'
    if (row['PFR'] >= 1 and row['PFR'] <= 3):
        return '1-3'
    if (row['PFR'] >= 4 and row['PFR'] <= 5):
        return '4-5'
    if (row['PFR'] >= 6 and row['PFR'] <= 9):
        return '6-9'
    if (row['PFR'] >= 10 and row['PFR'] <= 12):
        return '10-12'
    if (row['PFR'] >= 13 and row['PFR'] <= 15):
        return '13-15'
    if (row['PFR'] >= 20 and row['PFR'] <= 25):
        return '20-25'
    if (row['PFR'] >= 30 and row['PFR'] <= 40):
        return '30-40'
    if (row['PFR'] >= 41 and row['PFR'] <= 56):
        return '41-56'
    if (row['PFR'] >= 60):
        return '60+'
    
data['PFRType'] = data.apply (lambda row: labelObs(row), axis=1)
```

```python
d = pd.DataFrame(data['PFRType'].value_counts())
d
```

## Fuel Distribution

```python
data.head()
```

```python
data['MBP10_CF002'] = np.where(data['EtylenFlow'] > 0.1, 0, data['MBP10_CF002'])
data['MBP10_CF003'] = np.where(data['EtylenFlow'] > 0.1, 0, data['MBP10_CF003'])
data['MBP10_CF004'] = np.where(data['EtylenFlow'] > 0.1, 0, data['MBP10_CF004'])
data['gasSum'] = data.MBP10_CF002 + data.MBP10_CF003 + data.MBP10_CF004
data.head()
```

```python
data['H2_calc'] = np.where(data['MBP10_CF003'] > 0.0, np.round((data.MBP10_CF003/data.gasSum)*100), 0)
data['NG_calc'] = np.where(data['MBP10_CF002'] > 0.0, np.round((data.MBP10_CF002/data.gasSum)*100), 0)
data['N2_calc'] = np.where(data['MBP10_CF004'] > 0.0, np.round((data.MBP10_CF004/data.gasSum)*100), 0)
```

```python
data.head()
```

```python
data['totalFuel'] = data.H2_calc+data.N2_calc+data.NG_calc
cols = ['H2_calc','N2_calc','NG_calc','totalFuel','EtylenFlow']
data[cols]
```

```python
data['H2_cat'] = data['H2_calc']
data['NG_cat'] = data['NG_calc']
data['N2_cat'] = data['N2_calc']
```

```python
data['H2_cat'] = data['H2_cat'].apply(lambda x: 70 if (x >= 66 and x <= 75) else x)
data['NG_cat'] = data['NG_cat'].apply(lambda x: 30 if x==33  else x)
data['H2_cat'] = data['H2_cat'].apply(lambda x: 30 if x==33  else x)
data['H2_cat'] = data['H2_cat'].apply(lambda x: 20 if (x >= 20 and x <= 25) else x)
data['NG_cat'] = data['NG_cat'].apply(lambda x: 80 if (x >= 75 and x <= 85) else x)
data['NG_cat'] = data['NG_cat'].apply(lambda x: 70 if (x >= 65 and x <= 74) else x)
```

```python
fuel = data.groupby(['H2_cat','NG_cat']).size().reset_index().rename(columns={0:'count'})
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(fuel)
```

```python
'''
cat F1:
H2 - NG
30 - 70
40 - 60
50 - 50

cat F2:

H2 - NG
70 - 30
100 - 0
'''

def labelObs(row):
    if (row['H2_cat'] == 0 and row['NG_cat'] == 0):
        return 'Ethlyne'
    if (row['H2_cat'] == 0 and row['NG_cat'] == 100):
        return 'NG'
    if (row['H2_cat'] == 20 and row['NG_cat'] == 80):
        return '20H280NG'
    if (row['H2_cat'] == 30 and row['NG_cat'] == 70):
        return 'F1'
    if (row['H2_cat'] == 40 and row['NG_cat'] == 60):
        return 'F1'
    if (row['H2_cat'] == 50 and row['NG_cat'] == 50):
        return 'F1'
    if (row['H2_cat'] == 70 and row['NG_cat'] == 30):
        return 'F2'
    if (row['H2_cat'] == 100 and row['NG_cat'] == 0):
        return 'F2'
data['FuelType'] = data.apply (lambda row: labelObs(row), axis=1)
```
